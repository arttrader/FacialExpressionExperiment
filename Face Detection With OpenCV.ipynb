```
# check opencv version
import cv2
# print version number
print(cv2.__version__)
```

```
# example of face detection with opencv cascade classifier
from cv2 import imread
from cv2 import CascadeClassifier
# load the photograph
pixels = imread(this_path + 'test1.jpg')
# load the pre-trained model
classifier = CascadeClassifier(this_path + 'haarcascade_frontalface_default.xml')
# perform face detection
bboxes = classifier.detectMultiScale(pixels, 1.05, 8)
# print bounding box for each detected face
for box in bboxes:
	# extract
	x, y, width, height = box
	x2, y2 = x + width, y + height
	# draw a rectangle over the pixels
	cv2.rectangle(pixels, (x, y), (x2, y2), (0,0,255), 1)
```
